## Exercise 1
Write a function to compute the below formualation with the use of map function
\begin{equation}
x^2 + y^2 - r^2
\end{equation}

then, print out the values of the equation at these below values of $x$, $y$ and $z$
```
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,7]
r = [1,1,3,3,4,5]

```

````python
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,7]
r = [1,1,3,3,4,5]

result = list(map(lambda a, b, c: a**2 + b**2 - c**2, x, y, r))

print(result)
````

## Exercise 2

Write a function that take input as a list of words and return to a list of integers representing the lengths of the correponding words.

Write it in three different ways:
1. using a for-loop,
2. using function `map()`, and
3. using list comprehensions.

````python
def word_lengths_for(words):
    lengths = []
    for w in words:
        lengths.append(len(w))
    return lengths

def word_lengths_map(words):
    return list(map(len, words))

def word_lengths_comp(words):
    return [len(w) for w in words]
````

````python
words = ["apple", "banana", "kiwi"]
print(word_lengths_for(words))
print(word_lengths_map(words))
print(word_lengths_comp(words))
````

## Exercise 3

Implement a program to make a list whose elements are double string of a string `re` (i.e., `rere`), where `re` is the first two characters of approriate elements in this list:

`['real', 'rock', 'realize', 'rocket', 'red', 'eraser' , '7', 8 , 9, 10]`.

## Exercise 4

Write a function `max_in_list()` that takes a list of numbers and returns the largest one using `reduce()` in lib `functools`

````python
from functools import reduce

def max_in_list(numbers):
    return reduce(lambda a, b: a if a > b else b, numbers)
````

````python
nums = [3, 7, 2, 9, 5, 10, 1]
print(max_in_list(nums))
````

## Exercise 6
Implement parallel processing using `map()` to apply a function across multiple elements of an iterable concurrently, and then use `reduce()` to aggregate the results.

````python
from concurrent.futures import ProcessPoolExecutor
from functools import reduce

def square(x):
    return x * x

def parallel_square_and_sum(numbers):
    with ProcessPoolExecutor() as executor:
        results = list(executor.map(square, numbers))

    total = reduce(lambda a, b: a + b, results)
    return total

if __name__ == "__main__":
    nums = [1, 2, 3, 4, 5]
    result = parallel_square_and_sum(nums)
    print("Sum of squares:", result)
````

## Exercise 7
Implement a MapReduce-style function to process data distributed across multiple nodes or workers.

````python
from concurrent.futures import ProcessPoolExecutor
from functools import reduce
from collections import Counter

# ---- Step 1: Define the map function ----
def map_function(text_chunk):
    """Map phase: split text into words and count locally."""
    words = text_chunk.lower().split()
    return Counter(words)

# ---- Step 2: Define the reduce function ----
def reduce_function(counter1, counter2):
    """Reduce phase: merge two counters."""
    return counter1 + counter2

# ---- Step 3: The MapReduce controller ----
def map_reduce(text_chunks):
    # --- MAP PHASE ---
    with ProcessPoolExecutor() as executor:
        mapped_results = list(executor.map(map_function, text_chunks))

    # --- REDUCE PHASE ---
    final_result = reduce(reduce_function, mapped_results)
    return final_result

# ---- Example usage ----
if __name__ == "__main__":
    # Imagine each string is data from a different node
    data_chunks = [
        "apple orange banana apple",
        "banana apple orange orange",
        "grape banana apple"
    ]

    result = map_reduce(data_chunks)
    print(dict(result))
````
